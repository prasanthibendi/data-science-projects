### Task:

Create a python script (file submitted must be .py file extension) to calculate the base price on a weekly basis of each product per geography as new column called "bpru" in the “python_test_data.csv” dataset file. The final output csv file should have 5 columns: geography_key, product_key, period, ppu and bpru.

This "bpru" column is short for "Base price per unit". The python script should create a new csv file with all 5 columns and its data. Note your python script should work on any csv dataset file so any transformations you wish to do to the "python_test_data.csv" needs to be in your python script and not in the csv file as your script may be tested on a much larger dataset. The bpru reflects the median price over a 7 week period.

This bpru should be calculated as the median value of: the current periods ppu, the previous 3 periods ppu and the following 3 periods (so median of 7 ppu values) for each product per geography per week. This means that each product's bpru will be specific to that geography and may be different per geography (ex product "abc" on "2021-02-06" could have a different ppu and bpru in AB vs ON). The ppu may change week over week so you will need to sort by the period column chronologically to calculate the correct bpru. Additionally the first 3 periods per product per geography must be equal to the 4th periods bpru and the last 3 periods per product per geography must be equal to the 4th last periods bpru.

Note the bpru must be per product per geography so product "abc" in "ON" should only take into consideration the "ppu" values for "abc" in "ON" and not any other geographies or products.



The **geography_key** column represents the province.

The **product_key** column represents the product.

The **period** column represents the week ending this date.

The **ppu** column is short for "Price Per Unit" and represents the price of that product for that week for that province.

```python
geo_products = data.groupby('geography_key')['product_key']
geo_products.head()
```

```python
# Convert 'period' column to datetime to facilitate sorting
data['period'] = pd.to_datetime(data['period'])

# Sort the dataframe by 'geography_key', 'product_key', and 'period'
data_sorted = data.sort_values(by=['geography_key', 'product_key', 'period'])
```

```python
data_sorted[:]
```

#### Observations:

The above data is sorted by **period** based on unique **geography_key** and **product_key** combinations

```python
# Define a function to calculate the median ppu over a 7-week window
def calculate_bpru(series, window_size=7):
    return series.rolling(window=window_size, center=True, min_periods=window_size).median()

# Group data by 'geography_key' and 'product_key' and apply the median calculation
data_sorted['bpru'] = data_sorted.groupby(['geography_key', 'product_key'])['ppu'].transform(calculate_bpru)

data_sorted[:]
```

#### Observations:

The above **bpru** column represents a rolling median value for a set of 7 periods for each combination of **geography_key** and **product_key**.

#### Note:

Here, for the first 3 and last 3 values of **bpru**, it considered the available records (<7) for the median and returned `NaN`. 

Which we will fix below

```python
# Fix the first and last three periods
for key, group in data_sorted.groupby(['geography_key', 'product_key']):
    # Update the first three periods
    if len(group) > 3:
        data_sorted.loc[group.index[:3], 'bpru'] = group['bpru'].iloc[3]
        # Update the last three periods
        data_sorted.loc[group.index[-3:], 'bpru'] = group['bpru'].iloc[-4]
```

# Final Output

```python
# Show the updated data
data_sorted[:]
```

#### Observations:

In the above data, **bpru** value for the first 3 and last 3 periods are equal to **bpru** of 4th period and 4th last period respectively.

**Price Trends over time**: Lineplot 

```python
# Ensure 'period' column is a datetime object
data_sorted['period'] = pd.to_datetime(data_sorted['period'])

# Get the unique products
products = data_sorted['product_key'].unique()

for product in products:
    plt.figure(figsize=(10, 6))
    for geo, group in data_sorted[data_sorted['product_key'] == product].groupby('geography_key'):
        plt.plot(group['period'], group['ppu'], label=f"{geo}")
    plt.title(f"Price Trends Over Time for Product {product}")
    plt.xlabel("Date")
    plt.ylabel("Price Per Unit (ppu)")
    plt.legend(loc='upper left', bbox_to_anchor=(1, 1), ncol=2)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

#### Observations:

From the above line plots, we can determine if there are any price trends over time for each product across various geographies. 

With further market research, we can identify the root cause for these price trends.
